```python
img_augmentation_dir = '/tf/keras_neural_network/Mis_Tests/Faces/Mi_Cara'
target_count = 3000
```

```python
import os
image_paths = [os.path.join(img_augmentation_dir, f)
               for f in os.listdir(img_augmentation_dir)
               if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

current_count = len(image_paths)
print("Current images:", current_count)

if current_count == 0:
    raise ValueError("No images found in folder!")

# Data augmentation layer
augmenter = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.15),
    tf.keras.layers.RandomZoom(0.15),
    tf.keras.layers.RandomContrast(0.2),
])

# Loop until we reach target number of images
generated = 0
i = 0

while current_count + generated < target_count:
    img_path = image_paths[i % len(image_paths)]
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=(256, 256))
    img = tf.keras.preprocessing.image.img_to_array(img)
    img = tf.expand_dims(img, 0)

    # Apply augmentation
    aug_img = augmenter(img, training=True)[0].numpy().astype("uint8")

    # Save new file
    save_path = os.path.join(
        img_augmentation_dir,
        f"augmented_{current_count + generated:05d}.jpg"
    )
    tf.keras.preprocessing.image.save_img(save_path, aug_img)

    generated += 1
    i += 1

print(f"Generated {generated} new images.")
print("Done!")
```
